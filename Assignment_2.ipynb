```python
import numpy as np
from scipy.stats.kde import gaussian_kde
from matplotlib import pyplot as plt
def generate_data():
    return np.random.randn(10000)
```

```python
data=generate_data()
print data
```

```python
#pdf of the given data
kde=gaussian_kde(data)
dist_space = np.linspace( min(data), max(data))
plt.plot( dist_space, kde(dist_space) )
plt.show()
```

```python
def Probability(l):
    d={}
    for i in l:
        if d.has_key(i):
            d[i]+=1
        else:
            d[i]=1
    return d

def Expectation(l):
    e=0
    d=Probability(l)
    for x in d.keys():
        e+=(x*d[x]/len(d))
    return e
    
```

```python
#Normalization
normalized_data=[]
min_list=min(data)
max_list=max(data)
r=max_list-min_list
for d in data:
    normalized_data.append((d-min_list)/r)
print min_list
print max_list
        
```

```python
#pdf of the normalized data
kdeN=gaussian_kde(normalized_data)
dN = np.linspace( min(normalized_data), max(normalized_data))
plt.plot( dN, kdeN(dN) )
plt.show()
```

```python
#Expectation of the normalized data
print "Mean using inbuit function: "+str(np.mean(normalized_data)) #Just to verify
print "Standard Deviation: "+str(np.std(normalized_data))
print "Expectation using own function: "+str(Expectation(normalized_data))
```

```python
#Standardization:
standardized_data=[]
mean=np.mean(data)
std=np.std(data)
print mean #<- Expectation of non standardized data 
print std
for d in data:
    standardized_data.append(((d-mean)/std))
```

```python
#pdf of the standardized data
#Similar to the pdf of the non-standardized data
kdeS=gaussian_kde(standardized_data)
dS = np.linspace( min(standardized_data), max(standardized_data))
plt.plot( dS, kdeS(dS) )
plt.show()
```

```python
#Expectation of the standardized data
print np.std(standardized_data)
print Expectation(standardized_data) #which is very very close to 0
```
